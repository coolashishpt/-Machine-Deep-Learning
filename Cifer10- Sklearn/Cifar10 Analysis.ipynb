```python
import pandas as pd
import numpy as np
```

```python
x_train = images_train.reshape(50000,3072)
x_test = images_test.reshape(10000,3072)
```

```python
x_train.shape
```

```python
pd.DataFrame(x_train)
```

```python
from sklearn.decomposition import PCA
```

```python
pca1 = PCA()
pca1.fit(x_train)
```

```python
pca1.components_.shape
```

```python
# Finding best n_components value for reducing the speed.
k  = 0
total = 0
while total < 0.91:
    total += pca1.explained_variance_ratio_[k]
    k += 1
k
```

```python
pca2 = PCA(n_components=k, whiten=True)
x_transform = pca2.fit_transform(x_train)
x_transform.shape
```

```python
x_approx = pca2.inverse_transform(x_transform)
x_approx.shape
```

```python
x_approx = x_approx.reshape((50000, 32, 32, 3))
```

```python
x_train_pca = x_transform
x_test_pca = pca2.transform(x_test)
```

```python
fig = plt.figure(figsize=(8,8))
for i in range(64):
    ax = fig.add_subplot(8,8, 1 + i)
    ax.imshow(x_approx[i])
plt.show()
```

```python
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
```

```python
clf = SVC()
clf.fit(x_train_pca, cls_train)
```

```python
y_pred = clf.predict(x_test_pca)
```

```python
clf.score(x_train_pca, cls_train)
```
