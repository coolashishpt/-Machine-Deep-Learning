```python
document = []
for category in movie_reviews.categories():
    for fileid in movie_reviews.fileids(category):
        document.append((movie_reviews.words(fileid), category))
document[0:5]
```

```python
import string
from nltk.corpus import stopwords
stops = stopwords.words("english")
punctuations = list(string.punctuation)
stops_words = stops + punctuations
```

```python
from nltk.corpus import wordnet
from nltk import pos_tag
from nltk.stem import WordNetLemmatizer 
lemmatizer = WordNetLemmatizer() 
def get_simple_pos(tag):
    
    if tag.startswith("J"):
        return wordnet.ADJ
    
    elif tag.startswith("V"):
        return wordnet.VERB

    elif tag.startswith("N"):
        return wordnet.NOUN

    elif tag.startswith("R"):
        return wordnet.ADV

    else:
        return wordnet.NOUN
```

```python
def clean_data(document):
    output = []
    for word in document:
        if word.lower() not in stops_words:
            pos = pos_tag([word])
            lem_zer = lemmatizer.lemmatize(word, pos = get_simple_pos(pos[0][1]))
            output.append(lem_zer.lower())
    return output    
```

```python
new_documents = [(clean_data(word), category) for word, category in document]
```

```python
new_documents[0]
```

```python
training_data = new_documents[0:1500]
testing_data = new_documents[1500:]
```

```python
all_word = []
for i in training_data:
    all_word += i[0]
    
```

```python
import nltk
freq = nltk.FreqDist(all_word)
common = freq.most_common(3000)
features = [i[0] for i in common]
```

```python
features
```

```python
def get_feature_dic(words):
    current_feature = {}
    word_set = set(words)
    for  i in features:
        current_feature[i] = i in word_set
    return current_feature
```

```python
x_train = [(get_feature_dic(doc), category) for doc, category in training_data]
x_test = [(get_feature_dic(doc), category) for doc, category in testing_data]
```

```python
len(x_train), len(x_test)
```

```python
from nltk import NaiveBayesClassifier
clf = NaiveBayesClassifier.train(x_train)
```

```python
nltk.classify.accuracy(clf, x_test)
```

```python
clf.show_most_informative_features(15)
```
